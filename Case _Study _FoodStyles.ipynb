```python
review_columns=data[["excellent", "very_good", "average", "poor", "terrible", "total_reviews_count", "reviews_count_in_default_language"]]
rating_columns=data[["food", "service", "value", "atmosphere", "avg_rating"]]
```

Looking at the table above, we can see that there are not very high correlations between the variables.

```python
vegetarian_friendly=data[data['vegetarian_friendly']=='Y']

non_friendly=data[data['vegetarian_friendly']=='N']

print(vegetarian_friendly['avg_rating'].mean())
print(non_friendly['avg_rating'].mean())
```

```python
data[['low_limit','high_limit']] = data['price_range'].str.split('-',expand=True)
data['high_limit'] = data['high_limit'].str.replace('€', '', regex=True)
data['high_limit'] = data['high_limit'].str.replace('CHF\xa017', '', regex=True)
data['high_limit'] = data['high_limit'].str.replace("'",'', regex=True)
data['high_limit'] = data['high_limit'].str.replace('CHF', '', regex=True)
data['high_limit'] = data['high_limit'].str.replace(",",'', regex=True)
```

```python
data.high_limit=data.high_limit.replace(r'^\s*$', np.nan,regex=True)
# # data.groupby(['cuisines'])['high_limit'].mean()
data['high_limit']=data['high_limit'].fillna(0)
data['high_limit']=data['high_limit'].astype(float)
data.groupby(['cuisines'])['high_limit'].mean().nlargest(10)
```

When we group the cuisine types and take the average of the highest prices of these groups, we see that the restaurant selling Barbecue, European, Asian, Middle Eastern, Albanian, Bangladeshi, Afghani dishes has the highest average price.

```python
capital['Area in km2 '] = capital['Area in km2 '].str.replace(",",'.', regex=True)
capital['Area in km2 ']=capital['Area in km2 '].astype(float)
capital['population_km']=capital['Population']/capital['Area in km2 ']
```

```python
capital.sort_values(by='population_km',ascending=False)
```

```python
capital.rename(columns = {'City':'city'}, inplace = True)
merged=capital.merge(data, on='city', how='left')
out=(merged.groupby(['Country', 'city']).gluten_free.value_counts(normalize=True))
out_1=(merged.groupby(['Country', 'city']).gluten_free.value_counts().sort_values(ascending=False))
  
```

As a result of the analysis I made, I first compared the cities where the number of gluten-free restaurants is low with the cities where the population is high, and I came to the decision that more gluten-free restaurants could be opened in these cities. If the age distribution of the people living in the countries were also known, for example, since it is known that celiac disease, which is a disease, occurs more frequently in the 30-40 age range, I could comment that restaurants can be opened for cities with more people in this age range. And as a new variable, I calculated the number of people per square kilometer. If the number of people per square kilometer is high and the number of gluten-free restaurants is low, I think we can open restaurants in these cities. As a result, I identified 10 cities using this approach. These are Paris, Rome, Madrid, Milan, Prague, Stockholm, Lisbon, Zurich and Vienna.
